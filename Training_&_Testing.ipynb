```python
import pandas as pd
import numpy as np
```

```python
path = "/content/Tweet_Preprocessing.csv"
TWEET_DATA = pd.read_csv(path, usecols=["sentimen", "tweet_normalized"])
TWEET_DATA.columns = ["lable", "tweet"]
TWEET_DATA.lable = TWEET_DATA.lable.str.capitalize()
TWEET_DATA
```

# CONVERT STRING TO LIST

```python
# convert list formated string to list
import ast

def convert_text_list(texts):
    texts = ast.literal_eval(texts)
    return [text for text in texts]

TWEET_DATA["tweet_list"] = TWEET_DATA["tweet"].apply(convert_text_list)
pd.DataFrame(TWEET_DATA[["tweet_list", "lable"]])
```

## Split Dataset

```python
training_data = TWEET_DATA.sample(frac=0.8, random_state=50).reset_index()
training_data = training_data.drop(columns='index')
testing_data = TWEET_DATA.drop(training_data.index).reset_index()
testing_data = testing_data.drop(columns='index')
```

## Split Pos Neg Dataset

```python
training_data_pos = training_data[['lable', 'tweet_list']]
training_data_pos = training_data_pos[training_data_pos.lable == 'Positif'].reset_index()
training_data_pos = training_data_pos.drop(columns='index')
training_data_pos
```

```python
training_data_neg = training_data[['lable', 'tweet_list']]
training_data_neg = training_data_neg[training_data_neg.lable == 'Negatif'].reset_index()
training_data_neg = training_data_neg.drop(columns='index')
training_data_neg
```

**Menghitung Bobot TF**

```python
def calc_TF(document):
    # Counts the number of times the word appears in review
    TF_dict = {}
    for term in document:
        if term in TF_dict:
            TF_dict[term] += 1
        else:
            TF_dict[term] = 1
    for term in TF_dict:
        TF_dict[term] = TF_dict[term] / len(document)
    return TF_dict

training_data_pos["TF_dict"] = training_data_pos['tweet_list'].apply(calc_TF)
training_data_neg["TF_dict"] = training_data_neg['tweet_list'].apply(calc_TF)
training_data["TF_dict"] = training_data['tweet_list'].apply(calc_TF)
training_data_pos
```

```python
def call_tf(query):
    call_tf = {}
    index = query.index
    for i in index:
        for term in query[i]:
            if term in call_tf:
                call_tf[term] += query[i][term]
            else:
                call_tf[term] = query[i][term]
    return call_tf
```

```python
TF_Pos = call_tf(training_data_pos["TF_dict"])
df_TF_Pos = pd.DataFrame(TF_Pos.items(), columns=["Istilah", "TF-Positif"])

TF_Neg = call_tf(training_data_neg["TF_dict"])
df_TF_Neg = pd.DataFrame(TF_Neg.items(), columns=["Istilah", "TF-Negatif"])
```

```python
df_TF_Pos
```

**Menghitung Bobot DF**

```python
def calc_DF(tfDict):
    count_DF = {}
    # Run through each document's tf dictionary and increment countDict's (term, doc) pair
    for document in tfDict:
        for term in document:
            if term in count_DF:
                count_DF[term] += 1
            else:
                count_DF[term] = 1
    return count_DF
```

```python
DF = calc_DF(training_data["TF_dict"])
```

**Menghitung Bobot IDF**

```python
def calc_IDF(__n_document, __DF):
    # print("Menghitung IDF\n")
    IDF_Dict = {}
    d = {}
    for term in __DF:
        IDF_Dict[term] = np.log10(__n_document / (__DF[term]))
    return IDF_Dict
```

```python
# Stores the idf dictionary
n_document = len(training_data)
IDF = calc_IDF(n_document, DF)
```

```python
def rumus_IDF(__n_document, __DF):
    IDF_Dict = {}
    d = {}
    df = {}
    rumus = {}
    for term in __DF:
        IDF_Dict[term] = np.log10(__n_document / (__DF[term]))
        d[term] = __n_document
        df[term] = __DF[term]
        rumus[term] = 'log({}/{})'.format(round(d[term], 6), round(df[term], 4)) 
    IDF_Dict = pd.DataFrame({ 'D':pd.Series(d), 'DF':pd.Series(df), 'Rumus IDF = log(D/DF)':pd.Series(rumus), 'IDF':pd.Series(IDF_Dict)}).reset_index()
    return IDF_Dict
```

```python
df_IDF = rumus_IDF(n_document, DF)
df_IDF = df_IDF.rename(columns={'index': 'Istilah'})
df_IDF
```

**Total IDF**

```python
print('Total IDF = ', df_IDF['IDF'].sum())
```

**Menghitung Bobot TFIDF**

```python
#calc TF-IDF
def calc_TF_IDF(TF):
    TF_IDF_Dict = {}
    #For each word in the review, we multiply its tf and its idf.
    for key in TF:
        TF_IDF_Dict[key] = TF[key] * IDF[key]
    return TF_IDF_Dict
```

TFIDF Positif Negatif

```python
training_data_pos["TF-IDF_dict"] = training_data_pos["TF_dict"].apply(calc_TF_IDF)
training_data_pos
```

```python
training_data_neg["TF-IDF_dict"] = training_data_neg["TF_dict"].apply(calc_TF_IDF)
training_data_neg
```

# TF-IDF Dataframe

```python
def rumus_TF_IDF(__tf, __idf):
  __TF = {}
  __IDF =  {}
  rumus_TF_IDF = {}
  for key in __tf:
    __TF[key] = __tf[key]
    __IDF[key] = __idf[key]
    rumus_TF_IDF[key] = '{} * {}'.format(round(__tf[key], 6), round(__idf[key], 4)) 

  rumus_TF_IDF = pd.DataFrame({'TF':pd.Series(__TF), 'IDF':pd.Series(__IDF), 'Rumus':pd.Series(rumus_TF_IDF)}).reset_index()
  return rumus_TF_IDF 
```

```python
def fix_tf_idf(query):
    tf_idf_ = {}
    kemunculan = {}
    index = query.index
    for i in index:
        for term in query["TF-IDF_dict"][i]:
          if term in tf_idf_:
                tf_idf_[term] += query["TF-IDF_dict"][i][term]
                kemunculan[term] += 1
          else:
                tf_idf_[term] = query["TF-IDF_dict"][i][term]
                kemunculan[term] = 1
    tf_idf_ = pd.DataFrame({'Kemunculan':pd.Series(kemunculan), 'TF-IDF':pd.Series(tf_idf_)}).reset_index()
    return tf_idf_
```

**TF-IDF POSITIF**

```python
fix_tfidf_pos = fix_tf_idf(training_data_pos)
fix_tfidf_pos.columns = ['Istilah Positif', 'Kemunculan', 'TF-IDF']
rumus = rumus_TF_IDF(TF_Pos, IDF)
fix_tfidf_pos.insert(2, 'TF', rumus['TF'])
fix_tfidf_pos.insert(3, 'IDF', rumus['IDF'])
fix_tfidf_pos.insert(4, 'TF-IDF=TF*IDF', rumus['Rumus'])
fix_tfidf_pos
```

**TFIDF NEGATIF**

```python
fix_tfidf_neg = fix_tf_idf(training_data_neg)
fix_tfidf_neg.columns = ['Istilah Negatif', 'Kemunculan', 'TF-IDF']
rumus = rumus_TF_IDF(TF_Neg, IDF)
fix_tfidf_neg.insert(2, 'TF', rumus['TF'])
fix_tfidf_neg.insert(3, 'IDF', rumus['IDF'])
fix_tfidf_neg.insert(4, 'TF-IDF=TF*IDF', rumus['Rumus'])

fix_tfidf_neg
```

# Testing Data

```python
testing_data
```

```python
def calc_probability(uji, kelas, idf):
    calc_prob = {}
    for term in uji:
        if term in kelas.index:
          calc_prob[term] = (kelas['TF-IDF'][term]+1)/(kelas['TF-IDF'].sum()+idf['IDF'].sum())
        else:
          calc_prob[term] = (0+1)/(kelas['TF-IDF'].sum()+idf['IDF'].sum())
    return calc_prob
```

```python
kelas_pos = fix_tfidf_pos.set_index('Istilah Positif')
kelas_neg = fix_tfidf_neg.set_index('Istilah Negatif')
```

```python
testing_data['Cond Probability Positif'] = testing_data['tweet_list'].apply(calc_probability, kelas=(kelas_pos), idf=(df_IDF))
testing_data['Cond Probability Negatif'] = testing_data['tweet_list'].apply(calc_probability, kelas=(kelas_neg), idf=(df_IDF))
testing_data
```

```python
def Posterior_Probability(conditional_prob, prior_probability):
    posterior_prob = 1
    for key in conditional_prob:
        posterior_prob = posterior_prob*conditional_prob[key]
    posterior_prob = posterior_prob*prior_probability
    return posterior_prob
```

```python
total_train = len(training_data_pos)+len(training_data_neg)
prior_probability_pos = len(training_data_pos)/total_train
prior_probability_neg = len(training_data_neg)/total_train

testing_data['Post_Prob_Pos'] = testing_data['Cond Probability Positif'].apply(Posterior_Probability, prior_probability=(prior_probability_pos))
testing_data['Post_Prob_Neg'] = testing_data['Cond Probability Negatif'].apply(Posterior_Probability, prior_probability=(prior_probability_neg))
```

```python
testing_data
```

```python
testing_data['Prediksi'] = np.where(testing_data['Post_Prob_Pos'] >= testing_data['Post_Prob_Neg'], 'Positif', 'Negatif')
testing_data.lable = testing_data.lable.str.capitalize()
testing_data
```

```python
testing_data['Kesesuaian'] = np.where(testing_data['Prediksi'] == testing_data['lable'], 'Sesuai', 'Tidak Sesuai')
testing_data
```

```python
sesuai = len(testing_data[testing_data.Kesesuaian == 'Sesuai'])
total = len(testing_data)
akurasi = (sesuai/total)*100
```
